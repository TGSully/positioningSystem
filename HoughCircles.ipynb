```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread("/Users/beard0/Desktop/testball.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hsvImage= cv2.cvtColor(image ,cv2.COLOR_RGB2HSV)
plt.imshow(image)

```

```python
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
gray_blur = cv2.GaussianBlur(gray, (3,3), 0)


h = hsvImage[:,:,0]
hblur = cv2.GaussianBlur(h, (3,3), 0)
plt.imshow(hblur, cmap="gray")

threshBlur =  cv2.inRange(h, 75, 90)
plt.imshow(threshBlur, cmap='gray')
```

```python
## for drawing circles on
circles_im = np.copy(image)

## TODO: use HoughCircles to detect circles
# right now there are too many, large circles being detected
# try changing the value of maxRadius, minRadius, and minDist
circles = cv2.HoughCircles(threshBlur, cv2.HOUGH_GRADIENT, 1, 
                           minDist=500,
                           param1=70,
                           param2=11,
                           minRadius=60,
                           maxRadius=100)

# convert circles into expected type
circles = np.uint16(np.around(circles))
# draw each one
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(circles_im,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    ###cv2.circle(circles_im,(i[0],i[1]),2,(0,0,255),3)
    
plt.imshow(circles_im)

print('Circles shape: ', circles.shape)
    
```
